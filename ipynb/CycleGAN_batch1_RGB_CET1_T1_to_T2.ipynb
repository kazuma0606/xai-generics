```
#fake T1 epoch 200
import cv2
import pathlib
import glob
import re
import os
import numpy as np
import matplotlib.pyplot as plt
import nrrd
from PIL import Image


#fake_A: plain T1
#real_B: CE T1

def img_search(case, name):
  img_list = []
  images = []
  img_resize = []

  p_dir = pathlib.Path('/content/drive/My Drive/Colab Notebooks/CycleGAN_Project/pytorch-CycleGAN-and-pix2pix/') #home directory
  path = p_dir / 'results' / case / 'T1W2CE_batch_1_cyclegan/test_200/images'

  for i in range(0,155):
    s = '*_'+str(i)+name
    img_list.append([j for j in path.glob(s)])
  for i in range(len(img_list)):
    images.append(cv2.cvtColor(cv2.imread(str(img_list[i][0])), cv2.COLOR_BGR2RGB))
  for i in range(len(images)):
    img_resize.append( cv2.resize(images[i], (240,240)))

  return img_resize
```

```
def Imge_list_to_RGB(img):
  img = Image.fromarray(img)
  r, g, b= img.split()
  return r, g, b
```

```
import numpy as np
img_black = np.zeros((240, 240), np.uint8)
```

```
img_black.shape
```

```
pil_image = Image.fromarray(img_black)
```

```
list[100].stem[:9]
```

```
#fake_A: plain T1 '_fake_A.png'
#real_B: CE T1 '_real_B.png'

out_dir = pathlib.Path('/content/drive/My Drive/Colab Notebooks/CycleGAN_Project/pytorch-CycleGAN-and-pix2pix/datasets/T1W2T2W/RGB_CET1-T1_to_T2/trainB/')

j= 0
while j < 155:
  patient = list[j].stem[:9]
  r,g,b = [], [], []
  plain_T1_g = []
  CE_T1_r = []
  out = []
  plain_T1 = []
  CE_T1 = []

  plain_T1 = img_search(patient, '_fake_A.png')

  for i in range(len(plain_T1)):
    _,g,_ = Imge_list_to_RGB(plain_T1[i])
    plain_T1_g.append(g)

  CE_T1 = img_search(patient, '_real_A.png')

  for i in range(len(CE_T1)):
    r,_,_ = Imge_list_to_RGB(CE_T1[i])
    CE_T1_r.append(r)

  for i in range(len(CE_T1_r)):
    out = Image.merge('RGB', (CE_T1_r[i], plain_T1_g[i], pil_image))
    out.save(str(out_dir)+ "/" + patient + '_RGB_CET1-T1' +'_img_' + str(i) +'.png')

  j+=1
  print("done, case:"+patient)
```
